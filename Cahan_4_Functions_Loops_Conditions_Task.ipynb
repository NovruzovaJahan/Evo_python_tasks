Write a function that returns weather condition as string, based on given argument (not input) as integer value:

temp < 0 → "Freezing"

0–20 → "Cold"

21–30 → "Warm"

30 → "Hot"

```python
def weather(temp):
  if temp < 0:
    print("Freezing")
  elif temp <= 20:
    print("Cold")
  elif temp <= 30:
    print("Warm")
  else:
    print("Hot")
weather(33)
```

Create a function that checks the lenght of word given as argument. If it is longer than 8 characters, prints out "acceptable", if not, prints "failed".

```python
def check_length(x):
  if len(x) > 8:
    print("accestable")
  else:
    print("failed")
check_length("salam Azerbaycan")

# or

def check_length_2(x):
  return 'acceptable' if len(x) > 8 else 'failed'
check_length_2("salam")
```

```python
num_list = [1, 56, 3, 23, 12, 91, 53, 21, 8, 4, 15, 73, 90, 35, 47, 82, 7, 5]
```

Write a function, where you get an integer number from user and return the square root of that number, if it is perfect square root, if not, print "it is not a square of any number".

```python
import math

def find_SquareRoot():
  number = int(input("Enter a number to find square root: "))
  root = math.sqrt(number)
  if root.is_integer():
    return root
  else:
    print("It is not a square of any number")

find_SquareRoot()
```
